```python

#Import PyPDF2 Library if not install type pip install PyPDF2 in CMD to install this library
import PyPDF2
#Import gtts Library Google text to speech 
from gtts import gTTS
#Open PDF File
PdfObj=open('PDF/Upwork Proposal Sample.pdf',"rb")

#Read PDF File
pdfReader=PyPDF2.PdfFileReader(PdfObj)


MyText=""

# Extract Page Number
for PageNum in range(pdfReader.numPages):
    PageObj=pdfReader.getPage(PageNum)
#Extract Content on Pages
    MyText+=PageObj.extractText()
# Print Text
print(MyText)
PdfObj.close()

#Convert text to Audio File
#You can change langauges by changing lang ="code" in the text below.Languages code can be found on this website https://cloud.google.com/text-to-speech/docs/voices
TextToSpeech=gTTS(text=MyText,lang='en-IN')
TextToSpeech.save("Output.mp3")


```
